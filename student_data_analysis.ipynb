```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('student_scores.csv')
```

```python
df = df.drop('Unnamed: 0', axis = 1)
```

```python
df['WklyStudyHours'] = df['WklyStudyHours'].str.replace('5 - 10	', '5-10')
```

```python
gb = df.groupby('ParentEduc').agg({'MathScore': 'mean', 'ReadingScore':'mean', 'WritingScore':'mean'})
```

```python
print(gb)
```

```python
plt.figure(figsize=(6,6))
plt.title("Relationship between parent degree and student score")
sns.heatmap(gb, annot=True)
plt.show()
```

```python
# From the above chart we have analysed that parents education have a high impact on student scores
```

```python
gb = df.groupby('ParentMaritalStatus').agg({'MathScore': 'mean', 'ReadingScore':'mean', 'WritingScore':'mean'})
print(gb)
```

```python
plt.figure(figsize=(6,6))
plt.title("Relationship between parent marital status and student score")
sns.heatmap(gb, annot=True)
plt.show()
```

```python
# From the above chart we have concluded that marks have no impact of marrital status on student marks
```

```python
plt.figure(figsize=(12,6))
plt.title("Detecting outlier of mathscore")
sns.boxplot(data = df, x='MathScore')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title("Detecting outlier of reading score")
sns.boxplot(data = df, x='ReadingScore')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title("Detecting outlier of writing score")
sns.boxplot(data = df, x='WritingScore')
plt.show()
```

```python
# From above analysis we conclude that students are weak in maths in comparision with reading and writing
```

```python
print(df['EthnicGroup'].unique())
```

```python
groupA = df.loc[(df['EthnicGroup']) == 'group A'].count()
groupB = df.loc[(df['EthnicGroup']) == 'group B'].count()
groupC = df.loc[(df['EthnicGroup']) == 'group C'].count()
groupD = df.loc[(df['EthnicGroup']) == 'group D'].count()

l = ['group A', 'group B','group C','group D']
plt.title('The distribution of the ethnic group on student data')
mlist = [groupA['EthnicGroup'], groupB['EthnicGroup'],groupC['EthnicGroup'],groupD['EthnicGroup']]
plt.pie(mlist, labels = l, autopct = '%1.2f%%')
plt.show()
```

```python
# From above diagram we have concluded as teh group c has the most portion of ethnic group
```
